```python
# Importamos pandas
import pandas as pd
import ast
#Leemos el archivo csv y obtenemos el dataframe1
df_movies = pd.read_csv("../Data/movies_dataset.csv")
```

```python
#Visualizamos la cabecera del dataframe 
df_movies.head()
```

```python
#Mostramos la información del dataframe
df_movies.info()
```

```python
# Elimino columnas no necesarias
df_movies.drop(columns=['adult','homepage','imdb_id','original_title','poster_path','video'],inplace=True)

# Seteo el indice
df_movies.reset_index(drop=True, inplace=True)
```

```python
# Muestro las columnas
df_movies.columns
```

```python
# Muestro los Nulos
df_movies.isnull().sum()
```

```python
# Convertir la columna 'release_date' al formato de fecha
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], errors='coerce')
```

```python
# Los valores nulos del campo release date deben eliminarse.
df_movies.dropna(subset=['release_date'],inplace=True)
```

```python
# Seteo el indice
df_movies.reset_index(drop=True, inplace=True)
```

```python
# Extraer el año y crear una nueva columna 'release_year'
df_movies['release_year'] = df_movies['release_date'].dt.year.astype('Int64')
```

```python
# Verifico los nulos
df_movies.isnull().sum()
```

```python
def parse_dict_string(value):
    if pd.isna(value):
        return None
    else:
        return ast.literal_eval(value)

# Para aplicar parse_dict_string para pasar de diccionario a string
df_movies['genres'] = df_movies['genres'].apply(parse_dict_string)

# Función para extraer los valores de "name" de cada registro
def extract_name(record):
    return [item['name'] for item in record]

# Aplicar la función a la columna 'genres' y expandir los resultados en una nueva columna 'name_genres'
df_movies['name_genres'] = df_movies['genres'].apply(lambda x: extract_name(x))
```

```python
df_movies
```

```python
#Cambio a float el tipo de campo popularity
df_movies['popularity'] = df_movies['popularity'].astype('float')
```

```python
# Busqueda de duplicados
duplicados = df_movies['id'].value_counts() > 1
df_movies[df_movies['id'].isin(duplicados[duplicados].index)].sort_values(by=['id', 'popularity'],ascending=True)
```

```python
#Eliminación de Duplicados
df_movies = df_movies.drop_duplicates(subset=['id'], keep='last').sort_values(by=['id', 'popularity'], ascending=True)
```

```python
# Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0.
df_movies['revenue'] = df_movies['revenue'].fillna(0)
df_movies['budget'] = df_movies['budget'].fillna(0)
```

```python
# Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas 
# revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

# Convertir la columna 'budget' a tipo float
df_movies['budget'] = df_movies['budget'].astype(float)

# Calcular la columna "return" con la división de "revenue" entre "budget", teniendo en cuenta los casos nulos y 0
df_movies['return'] = df_movies.apply(lambda row: row['revenue'] / row['budget'] if row['revenue'] and row['budget'] else 0, axis=1)
```

```python
# Eliminacion de columnas que no se usaran
df_movies.drop(columns=['belongs_to_collection','original_language','production_companies','production_countries','runtime','spoken_languages','status','tagline'],inplace=True)

# Seteo el indice
df_movies.reset_index(drop=True, inplace=True)
```

```python
# Convertir id a entero
df_movies['id'] = df_movies['id'].astype(int)
```

```python
import ast
# Funcion para desanidar una columna añadiendo un prefijo
def desanidar_columna_con_prefijo(df, column, prefix):
    # Convertir la columna a una lista de diccionarios
    df[column] = df[column].apply(ast.literal_eval)
    # Crear un nuevo DataFrame con la data desanidada
    df_desanidado = df.explode(column)
    # Resetear el indice
    df_desanidado = df_desanidado.reset_index(drop=True)
    # Convertir los diccionarios a una columna individual
    df_desanidado = pd.concat([df_desanidado.drop([column], axis=1), df_desanidado[column].apply(pd.Series)], axis=1)
    # Añadir un prefijo a las columnas
    df_desanidado = df_desanidado.add_prefix(prefix)
    # Renombrar la columna id con su nombre original
    df_desanidado = df_desanidado.rename(columns={f'{prefix}idpelicula': 'id'})
    return df_desanidado

# Renombrar la columna id 
df_credits = df_credits.rename(columns={'id': 'idpelicula'})
# Desanidar la columna 'cast' con el prefijo 'cast_'
cast_desanidada_conprefijo = desanidar_columna_con_prefijo(df_credits[['idpelicula', 'cast']], 'cast', 'cast_')

# Renombrar la columna id 
df_credits = df_credits.rename(columns={'id': 'idpelicula'})
# Desanidar la columna 'crew' con el prefijo 'crew_'
crew_desanidada_conprefijo = desanidar_columna_con_prefijo(df_credits[['idpelicula', 'crew']], 'crew', 'crew_')
```
